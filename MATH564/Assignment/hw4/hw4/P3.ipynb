```r
data_injury <- read.table("Table6.6.txt", 
                          head = TRUE,
                          sep = "\t")
#data_injury
```

```r
mod_injury1 <- lm(Y ~ N, data_injury) 
print(summary(mod_injury1))
```

```r
res <- resid(mod_injury1) 
plot(fitted(mod_injury1), res) 
abline(0,0)
```

AS shown above the error is not heteroscedastic, and the ”E” assumption is violated.

(b)

```r
mod_injury2 <- lm(sqrt(Y) ~ N, data_injury) 
print(summary(mod_injury2))
```

Regression line equation: $\sqrt{Y}=\beta_{0}+\beta_{1} N+\varepsilon$ That is $\sqrt{Y}=1.17+11.86 N+\varepsilon$

```r
res <- resid(mod_injury2) 
plot(fitted(mod_injury2), res) 
abline(0,0)
```

Now there is no heteroscedasticity.
