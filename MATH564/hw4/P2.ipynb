```r
mod_bac2 <- lm(log(N_t) ~ t, data_bac) 
print(summary(mod_bac2))
```

Regression line equation:
$$\log \left(n_{t}\right)=\beta_{0}+\beta_{1} t+\varepsilon$$ That is $$ Log(n_{t} )= 5.97 -0.22t + \varepsilon$$
OR
$$n_{t}=\exp \left(\beta_{0}+\beta_{1} t\right)+\varepsilon$$ That is $$ n_{t}=\exp (5.97-0.22 t)+\varepsilon$$

```r
res <- resid(mod_bac2) 
plot(fitted(mod_bac2), res) 
abline(0,0)
```

The “L” assumption is no longer violated.
